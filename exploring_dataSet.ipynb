Problem Statement:
the goal is to load and explore the iris dataset. The aim is to understand the dataset structure, explore feature distributions and relationships and use data visualizations to identify any outliers or patterns.

```python
#importing libraries
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = sns.load_dataset("iris")  #iris dataset loaded
```

```python
print("Shape: ",df.shape)  #shape
print("Columns :", df.columns)  #column names
print(df.head())  #for the first 5 rows
```

```python
print(df.info())  #non-null count, data types
print(df.describe())   #statistics
print("Missing values: ",df.isnull().sum())  #missing values
```

```python
sns.scatterplot(data = df, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
sns.histplot(data = df, x='sepal_length', bins=20, kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x='sepal_width', bins=20, kde=True)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x='petal_length', bins=20, kde=True)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x='petal_width', bins=20, kde=True)
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.show()
```

```python
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    sns.boxplot(data=df, x='species', y=col)
    plt.title(f"{col} by Species")
    plt.show()
```

```python
#conclusion
#-Iris dataset is well-balanced
#-It has no missing values
#-Petal length and width prove to be the indicatiors of species
#-As per the box plot setosa has smaller petals, virginca generally has largest
#-Through data visualization we can see species seperation is clearhence classification models will perfrom well
```
